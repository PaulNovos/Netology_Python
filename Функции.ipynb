# Автосекретарь

```python
# Список документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# Список полок с номерами документов, которые на них хранятся
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

"""
Список команд:
"p" - узнать владельца документа по его номеру
"s" - по номеру документа узнать на какой полке он хранится
"l" - увидеть полную информацию по всем документам
"ads" - добавить новую полку
"ds" - удалить существующую полку из данных (только если она пустая)
"ad" - добавить новый документ в данные
"d" - удалить документ из данных
"m" - переместить документ с полки на полку
"q" - выход из программы
"""

# ------------- Блок функций---------

# Выводим владельца документа по номеру документа. Параметр doc_num - номер документа. Функция возвращает 1 в случае успеха.
def owner(doc_num):
  for doc in documents: # для каждого словаря doc в documents:
    if doc['number'] == doc_num: # если по ключу словаря number номер совпадает с переданным аттрибутом, то
      print(f"Владелец документа: {doc['name']}") # выводим результат
      return 1 # выход из функции
  print('Документ не найден в базе') # если номер не совпал

# Выводим номер полки по номеру документа. Параметр doc_num - номер документа. Функция возвращает номер полки в случае успеха.
def number_directory(doc_num):
  for k, v in directories.items(): # для каждого ключа k и его значения v в directories:
    if doc_num in v: # если номер документа в списке значений v, то:
      return k # возвращаем ключ = номер полки
  print('Документ не найден в базе') # если такого номера нет

# Выводим текущий список документов
def list_documents():
  for doc in documents: # для каждого словаря doc в documents:
    print(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {number_directory(doc['number'])}") # вывод. полку получаем из функции number_directory()

# Добавление полок. Параметр num_dir - номер полки
def add_directory(num_dir):
  if num_dir not in directories.keys(): # если номера добавляемой полки ещё нет в списке полок (ключи словаря directories), то:
    directories[num_dir] = [] # добавляем запись в словарь: ключ - номер полки, значение - пустой список
    print(f"Полка добавлена. Текущий перечень полок: {', '.join(directories.keys())}.") # номера полок - джойним в строку с разделителем ", " список ключей словаря directories 
  else: print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(directories.keys())}.") # если номер добавляемой полки уже в списке полок

# Удаление полок. Параметр num_dir - номер полки
def del_directory(num_dir):
  if num_dir not in directories.keys(): # если номера добавляемой полки нет в списке полок (ключи словаря directories), то:
    print(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories.keys())}.") # сообщаем об этом
  elif directories[num_dir] != []: # иначе, если список документов по ключу полки не пустой, то
    print(f"На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {', '.join(directories.keys())}.") # сообщаем об этом
  else: # если всё готово к удалению, то
    del directories[num_dir] # удаляем запись с ключом номера полки
    print(f"Полка удалена. Текущий перечень полок: {', '.join(directories.keys())}.") # выводим результат

# Добавление документа. Параметры: doc_number - номер, doc_type - тип, doc_owner - владелец, doc_dir - номер полки.
def add_document(doc_number, doc_type, doc_owner, doc_dir):
  for doc in documents: # для каждого словаря в documents:
    if doc['number'] == doc_number: # если по ключу словаря number номер совпадает с переданным аттрибутом, то
      print('Документ с таким номером уже существует. Текущий список документов:') # сообщаем, что номер уже есть
      list_documents()
      return 0 # выход из функции
  if doc_dir not in directories.keys(): # если номера полки для документа нет в списке полок (ключи словаря directories), то:
    print("Такой полки не существует. Добавьте полку командой ads. Текущий список документов:") # сообщаем об этом
    list_documents()
  else: # если с полкой порядок, то:
    documents.append({'type': doc_type, 'number': doc_number, 'name': doc_owner}) # в список documents добавляем новую запись-словарь с данными документа
    directories[doc_dir].append(doc_number) # в словарь directories по ключу с номером полки добавляем запись в список документов
    print('Документ добавлен. Текущий список документов:') # выводим результат
    list_documents()

# Удаление документа. Параметр doc_num - номер документа. Функция возвращает 1 в случае успеха.
def del_document(doc_num):
  for doc in documents: # для каждого словаря в documents:
    if doc['number'] == doc_num: # если по ключу словаря number номер совпадает с переданным аттрибутом, то
      documents.remove(doc) # удаляем из списка documents значение doc, т.е. весь соответствующий словарь
      directories[number_directory(doc_num)].remove(doc_num) # из словаря с полками удаляем по ключу с номером полки (получаем из функции, которая возвращает номер полки по номеру документа) значение номера докумена в соответствующем списке
      print('Документ удален. Текущий список документов:') # выводим результат
      list_documents()
      return 1 # выход из функции
  # если номер не совпал, то
  print('Документ не найден в базе. Текущий список документов:')
  list_documents()

# Перемещение документа с полки на полку. Параметры: doc_number - номер документа, doc_dir - номер целевой полки. Функция возвращает 1 в случае успеха.
def doc_move(doc_number, doc_dir):
  if doc_dir not in directories.keys(): # если номера целевой полки нет в списке полок (ключи словаря directories), то:
    print(f"Такой полки не существует. Добавьте полку командой ads. Текущий перечень полок: {', '.join(directories.keys())}.") # сообщаем об этом
    return 0 # выход из функции
  for doc in documents: # для каждого словаря в documents:
      if doc['number'] == doc_number: # если по ключу словаря number номер совпадает с переданным аттрибутом, то
        directories[doc_dir].append(doc_number)
        directories[number_directory(doc_number)].remove(doc_number) # из словаря с полками удаляем по ключу с номером полки (получаем из функции, которая возвращает номер полки по номеру документа) значение номера докумена в соответствующем списке
        print('Документ перемещен. Текущий список документов:') # выводим результат
        list_documents()
        return 1 # выход из функции
  # если номер не совпал, то
  print('Документ не найден в базе. Текущий список документов:')
  list_documents()


# ------------- Блок ввода команд ----------------

def main():
 while True:
  command = input('Введите команду: ').lower()
  print(command)
  if command == 'q':
    print('До свидания!')
    break
  elif command in ('p', 's', 'd'):
    doc_num = input('Введите номер документа: ')
    if command == 'p': 
      owner(doc_num)
    elif command == 's': 
      if number_directory(doc_num): # т.к. функция сама не выводит результат, а только возвращает номер полки, то если она вернула true-результат, то:
        print(f"Документ хранится на полке: {number_directory(doc_num)}") # выводим результат поиска
    elif command == 'd':
      del_document(doc_num)
  elif command == 'l':
    list_documents()
  elif command in ('ads', 'ds'):
    num_dir = input('Введите номер полки: ')
    if command == 'ads': 
      add_directory(num_dir)
    elif command == 'ds': 
      del_directory(num_dir)
  elif command == 'ad':
    doc_number = input('Введите номер документа: ')
    doc_type = input('Введите тип документа: ')
    doc_owner = input('Введите владельца документа: ')
    doc_dir = input('Введите полку для хранения: ')
    add_document(doc_number, doc_type, doc_owner, doc_dir)
  elif command == 'm':
    doc_number = input('Введите номер документа: ')
    doc_dir = input('Введите полку для хранения: ')
    doc_move(doc_number, doc_dir)
  else:
    print('Нет такой команды, попробуйте ещё раз!')

main()
```

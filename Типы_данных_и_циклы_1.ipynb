# Задание 1
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:

среднюю букву, если число букв в слове нечетное;

две средних буквы, если число букв четное.

```python
word = input('Введите слово: ')

if len(word) % 2 == 0: # если четное число букв
  print('Средние буквы: ', word[int(len(word)/2) - 1 : int(len(word)/2 + 1)]) # len(word)/2 - середина слова, "-1", т.к. индексы с 0, а "+1", т.к. правый индекс в срез не входит
else:
  print('Средняя буква: ', word[len(word)//2]) # при нечетном числе букв середина слова - это результат целочисленного деления на 2. С учетом первого индекса "0" это как раз нужная буква
```

# Задание 3
Мы делаем MVP dating-сервиса, и у нас есть список парней и девушек.

Выдвигаем гипотезу: лучшие рекомендации мы получим, если просто отсортируем имена по алфавиту и познакомим людей с одинаковыми индексами после сортировки! 

Но мы не будем никого знакомить, если кто-то может остаться без пары.

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

if len(boys) != len(girls):  # проверка на соответствие кол-ва элементов в списках
  print('Внимание, кто-то может остаться без пары!')
else:
  boys.sort() # сортируем списки по алфавиту
  girls.sort()
  print('Идеальные пары:')
  i = 0
  while i < len(boys): # выводим пары с одинаковыми индексами из обоих списков в цикле по длине одного из списков
    print(f'{boys[i]} и {girls[i]}')
    i += 1
```

# Задание 5 (необязательное)
Дан поток логов по количеству просмотренных страниц для каждого пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

```python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user7,4',
    '2018-03-29,user5,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user5,4',
    '2018-04-22,user5,8',
    '2018-05-03,user5,9',
    '2018-05-11,user6,11',
]

stream_user = [0, stream[0].split(',')[1]] #создаем список с уникальными именами пользователей и заносим первое имя из исходного списка. "0" в начале, чтобы создался список, а не одна строка.
page = 0 # создаем счетчик просмотренных страниц и инициализируем нулем

for log in stream:  # перебираем строчки в исходном списке
  page += int(log.split(',')[2])  # каждую строку разбиваем на список строк по ",". Значение числа страниц преобразуем в число и добавляем в счетчик.
  if log.split(',')[1] not in stream_user:  # проверяем имя пользователя на уникальность (если его ещё нет в списке пользователей, то:)
    stream_user.append(log.split(',')[1]) # добавляем уникальное имя пользователя в список имен

print(f'Среднее количество просмотров на уникального пользователя: {round(page / len(stream_user[1:]), 2)}')  # сред. кол-во просмотров - кол-во страниц / длину списка имен без первого "0" и округляем до 2-х знаков
```

# Задание 6 (необязательное)
Необходимо у пользователя запрашивать набор чисел разделенных пробелом. В результате в отсортированном порядке должны выводиться числа, которые повторяются в вводе более одного раза.

```python
# Ввод чисел по одному

number = input('Введите число (enter для завершения): ') # запрос первого числа

number_list = [] # создаем пустой список чисел

if number != '': # если что-то ввели
  number_list_double = [] # создаем пустой список чисел-дублей
  while number != '': # цикл "запрос числа - сортировка по спискам", пока не получим от пользователя "Enter"
      if int(number) in number_list and number not in number_list_double: # если число уже есть в списке чисел (дубль), но его ещё нет в списке дублей, то
        number_list_double.append(number) # добавляем число в список дублей
      number_list.append(int(number)) # добавляем введенное число в список чисел
      number = input('Введите число (enter для завершения): ') # запрашиваем новое число
  number_list_double.sort() # сортируем список дублей
  for n in number_list_double: # вывод чисел из отсортированного списка дублей
    print(n, end=' ')
else: # если изначально ничего не ввели
  print('Вы ничего не ввели')
```

Задание 6 (другой способ ввода)

```python
# Ввод сразу всех чисел в одном запросе

numbers = input('Введите числа (через пробел): ') # запрос чисел

if numbers != '': # если что-то ввели
  num_list = numbers.split(' ') # создаем список строк с числами, разбивая исходную строку по пробелам
  num_list_dbl = [] # создаем пустой список чисел-дублей
  for number in num_list: # для каждой строки с числом:
      if num_list.count(number) > 1 and int(number) not in num_list_dbl: # если количество вхождений строки (числа) в список больше 1 и такого числа ещё нет в списке дублей, то
        num_list_dbl.append(int(number)) # добавляем число в список дублей
  num_list_dbl.sort() # сортируем список дублей
  for n in num_list_dbl: # вывод чисел из отсортированного списка дублей
    print(n, end=' ')
else: # если изначально ничего не ввели
  print('Вы ничего не ввели')
```
